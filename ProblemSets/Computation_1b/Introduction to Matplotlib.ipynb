## Problem 1

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 1
def ran_normal(n):
    matrix = np.random.normal(size=(n,n))
    row_mean=np.mean(matrix, axis=1)
    var=np.var(row_mean)
    return var
```

```python
n=3
A = ran_normal(n)
print(A)
```

```python
num_eval=len(np.arange(100, 1001, 100))
results=np.zeros(num_eval)
for ii, n in enumerate(range(100, 1001, 100)):
    results[ii] = ran_normal(n)
```

```python
# plot the variance
plt.plot(np.arange(100, 1001, 100), results)
```

```python
x = np.linspace(0, 2*np.pi, 150)

ax1 = plt.subplot(221)
plt.axis([0, 2*np.pi, -2, 2])
ax1.set_title("sin(x)")
ax1.plot(x, np.sin(x), color="green")

ax2 = plt.subplot(222)
plt.axis([0, 2*np.pi, -2, 2])
ax2.set_title("sin(2x)")
ax2.plot(x, np.sin(2*x), color="red", linestyle="dashed")

ax3 = plt.subplot(223)
plt.axis([0, 2*np.pi, -2, 2])
ax3.set_title("2sin(x)")
ax3.plot(x, 2*np.sin(x), color="blue", linestyle="dashed")

ax4 = plt.subplot(224)
plt.axis([0, 2*np.pi, -2, 2])
ax4.set_title("2sin(2x)")
ax4.plot(x, 2*np.sin(2*x), color="magenta", linestyle="dotted")
plt.suptitle("Figures Problem 4");
```
