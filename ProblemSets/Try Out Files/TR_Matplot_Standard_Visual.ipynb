```python
fig, axes = plt.subplots(1, 2)
window = [x.min(), x.max(), y.min(), y.max()]

for ax, size in zip(axes, [10, 25]):
    ax.hexbin(x, y, gridsize=size, cmap='inferno')
    ax.axis(window)
    ax.set_aspect("equal")
    
    plt.show()
```
